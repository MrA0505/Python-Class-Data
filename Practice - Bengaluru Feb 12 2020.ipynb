```python
def convertval(var):
    total=0
    if '-' in var:
        var1=var.split('-')
        for a in var1:
            total=total+float(a)
        avg=total/len(var1)
        return avg
    else:
        return float(var)
```

```python
print(convertval('1200 - 1400'))
```

```python
print(convertval('2800'))
```

```python
house['sqr_ft_new']=house['total_sqft'].apply(convertval)
```

```python
def check_all(var):
    if var.isdigit()==False:
        if var.find('.')!=-1:
            pass
        else:
            print(var,end=' ')
   # print(var,end=' ')
```
