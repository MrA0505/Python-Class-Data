```python
#concate 
# It is used to combine more than one data frame or series
```

```python
import pandas as pd
```

```python
ser1=pd.Series([1,2,3,4])
ser2=pd.Series([4,5,6,7])
```

```python
con_ser=pd.concat([ser1,ser2])
```

```python
con_ser
```

```python
con_ser=pd.concat([ser1,ser2],ignore_index=True)
```

```python
con_ser
```

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3']
                 })
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                 'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7']
                 })
```

```python
df1
```

```python
df2
```

```python
df_res=pd.concat([df1,df2]) # here it is combined by index or axis=0
```

```python
df_res
```

```python
df_res1=pd.concat([df1,df2],axis=1)
```

```python
df_res1
```

```python
df3=pd.DataFrame({'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3']
                 },index=[1,2,3,4])
df4=pd.DataFrame({'A':['A4','A5','A6','A7'],
                 'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7']
                 },index=[1,2,5,6])
```

```python
df3
```

```python
df4
```

```python
df_res2=pd.concat([df3,df4])
```

```python
df_res2
```

```python
df_res3=pd.concat([df3,df4],axis=1)
```

```python
df_res3
```

```python
df_res3=pd.concat([df3,df4],axis=1,join='inner')
```

```python
df_res3
```
